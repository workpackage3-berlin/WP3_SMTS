```python
import pyxdf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 500)
```

```python
data, header = pyxdf.load_xdf('scripts/data/test1.xdf')
```

```python
events = []
ts = []
for stream in data: 
    print(stream)
    for series, stamp in zip(stream['time_series'], stream['time_stamps']):
        events.append(series)
        ts.append(stamp)

df = pd.DataFrame({'events': events, 
                   'timestamp': ts})    
```
